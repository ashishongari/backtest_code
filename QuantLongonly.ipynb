```python
correlation = port['nifty'].corr(port['portfolio'])
print(correlation)
```

```python
window =100

Roll_Max_portfolio = port['portfolio'].rolling(window, min_periods=1).max()
Roll_Max_nifty = port['nifty'].rolling(window, min_periods=1).max()

Drawdown_port = ((port['portfolio']/Roll_Max_nifty) - 1.0)*100
Drawdown_nifty = ((port['nifty']/Roll_Max_portfolio) - 1.0)*100
```

```python
# Next we calculate the minimum (negative) daily drawdown in that window.
# Again, use min_periods=1 if you want to allow the expanding window
Max_Daily_Drawdown_port = Drawdown_port.rolling(window, min_periods=1).min()
Max_Daily_Drawdown_nifty = Drawdown_nifty.rolling(window, min_periods=1).min()

# Plot the results
plt.figure(figsize=(20,8))
Drawdown_port.plot()
Drawdown_nifty.plot()
plt.title('Max Drawdown')
plt.show()
```
